```python
# Testing traffic
from tensorflow import keras
import tensorflow as tf
import traffic
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load model
model_1 = keras.models.load_model('model_final.h5')
model_2 = keras.models.load_model('model_final-2.h5')
```

```python
images_, labels_ = traffic.load_data('../jpeg')
images_ = np.array(images_) / 255.0
```

```python
plt.figure(figsize=(12,12))
for i in range(len(images_)):
    plt.subplot(1, len(images_), i+1)
    plt.imshow(images_[i])
    plt.axis("off")
    predictions = model_2(tf.expand_dims(images_[i], 0))
    plt.title("{}, {}".format(np.where(predictions == np.max(predictions))[1][0], '%.2f'%np.max(predictions)))
```
