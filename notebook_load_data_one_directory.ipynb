```python
list_images_gtsrb_small = list()

# Outer loop over directory gtsrb_small
for path in os.listdir(dir_path_gtsrb_small):
    if path[0] != '.':
        # Inner Loop over all files in labeled directory
        for path_inner in os.listdir("{}/{}".format(dir_path_gtsrb_small, path)):
            if path_inner[0] != ".":
                list_images_gtsrb_small.append(
                    cv.resize(cv.imread("{}/{}/{}".format(dir_path_gtsrb_small, path, path_inner)), (30, 30), interpolation=cv.INTER_CUBIC)
                )
```
